```python
import h2o
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from math import sqrt
from h2o.automl import H2OAutoML

# Initialize H2O
h2o.init()

# Load Apple's stock data from Yahoo Finance
apple_stock_data = yf.download('AAPL', start='2004-01-01', end='2024-01-01')
apple_stock_data.reset_index(inplace=True)

# Calculate additional features
apple_stock_data['SMA_20'] = apple_stock_data['Close'].rolling(window=20).mean()
apple_stock_data['SMA_50'] = apple_stock_data['Close'].rolling(window=50).mean()
apple_stock_data['EMA_20'] = apple_stock_data['Close'].ewm(span=20, adjust=False).mean()
apple_stock_data['EMA_50'] = apple_stock_data['Close'].ewm(span=50, adjust=False).mean()
apple_stock_data['RSI'] = 100 - (100 / (1 + (apple_stock_data['Close'].diff().apply(lambda x: max(x, 0)).rolling(window=14).mean() / 
                                    apple_stock_data['Close'].diff().apply(lambda x: max(-x, 0)).rolling(window=14).mean())))
apple_stock_data['MACD'] = apple_stock_data['Close'].ewm(span=12, adjust=False).mean() - apple_stock_data['Close'].ewm(span=26, adjust=False).mean()
apple_stock_data['Bollinger_Upper'] = apple_stock_data['SMA_20'] + (apple_stock_data['Close'].rolling(window=20).std() * 2)
apple_stock_data['Bollinger_Lower'] = apple_stock_data['SMA_20'] - (apple_stock_data['Close'].rolling(window=20).std() * 2)
apple_stock_data['Volume'] = apple_stock_data['Volume']

# Drop rows with NaN values
apple_stock_data.dropna(inplace=True)

# Select relevant features
features = ['Open', 'High', 'Low', 'Close', 'Volume', 'SMA_20', 'SMA_50', 'EMA_20', 'EMA_50', 'RSI', 'MACD', 'Bollinger_Upper', 'Bollinger_Lower']
target = 'Close'

# Split the data into train (80%) and test (20%) sets
train_size = int(len(apple_stock_data) * 0.7)
train_data = apple_stock_data[:train_size]
test_data = apple_stock_data[train_size:]

# Convert Pandas DataFrame to H2OFrame
train_h2o = h2o.H2OFrame(train_data)
test_h2o = h2o.H2OFrame(test_data)

# Define the features and target
x = features
y = target

# Train an H2O AutoML model
aml = H2OAutoML(max_models=5, seed=1, max_runtime_secs=3600)
aml.train(x=x, y=y, training_frame=train_h2o)

# Retrieve and display the leaderboard
lb = aml.leaderboard
lb_df = lb.as_data_frame()
print(lb_df[['model_id', 'rmse']])

# Calculate and print RMSE for each model
rmse_data = []
for model_id in lb_df['model_id']:
    model = h2o.get_model(model_id)
    preds = model.predict(test_h2o).as_data_frame()['predict']
    rmse = sqrt(mean_squared_error(test_data['Close'], preds))
    rmse_data.append({'model_id': model_id, 'rmse': rmse})

# Convert RMSE data to DataFrame
rmse_df = pd.DataFrame(rmse_data)
print(rmse_df)

# Verify that all models' predictions are on the same scale as the actual values
predictions = {}
for model_id in lb_df['model_id']:
    model = h2o.get_model(model_id)
    preds = model.predict(test_h2o).as_data_frame()
    predictions[model_id] = preds['predict']

# Convert predictions to DataFrame
predictions_df = pd.DataFrame(predictions)
predictions_df['Date'] = test_data['Date'].values
predictions_df['Actual'] = test_data['Close'].values

# Plot actual and predicted values
plt.figure(figsize=(14, 10))
plt.plot(predictions_df['Date'], predictions_df['Actual'], label='Actual', color='black', linewidth=2)
for model_id in lb_df['model_id']:
    plt.plot(predictions_df['Date'], predictions_df[model_id], label=model_id)
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.title('Predictions for All Models on the Test Set')
plt.legend()
plt.show()
```
